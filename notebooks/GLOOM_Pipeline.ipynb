```python
# Interactive volcano plot
import plotly.express as px

de = pd.read_csv('outputs/results/differential_expression_results.csv', index_col=0)
de['neg_log10_pval'] = -np.log10(de['pvalue'].clip(lower=1e-300))

fig = px.scatter(de, x='log2FoldChange', y='neg_log10_pval',
                 hover_name=de.index, opacity=0.5,
                 title='Volcano Plot — Differential Expression',
                 labels={'log2FoldChange': 'Log₂ Fold Change', 'neg_log10_pval': '-Log₁₀(p-value)'})
fig.show()
```
